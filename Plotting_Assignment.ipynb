```python
# YOUR CODE HERE
plt.hist(df['co2'],edgecolor='red',bins=range(0, 700, 35))
plt.xlabel('co2 amount')
plt.xticks(range(0, 630, 80))
plt.ylabel('co2 frequency')
plt.show()
```

```python
# YOUR CODE HERE
plt.hist(df['city'],edgecolor='red')
plt.title('city fuel Economy')
plt.xlabel('city')
plt.ylabel('frequency')
plt.show()
```

**The histogram illustrates the distribution of city fuel efficiency. The data exhibits an approximately normal distribution, with a central peak around 20, and the majority of vehicles falling within the 15-25 range. As fuel efficiency moves further from the center, the frequency of vehicles with lower or higher values diminishes. The graph highlights that most vehicles tend to have a city fuel economy near the average, while fewer vehicles achieve significantly higher or lower efficiency.**

```python
# YOUR CODE HERE
plt.hist(df['fuelType'],edgecolor='red')
plt.title('fuel type distribution ')
plt.show()
```

**Based on the histogram, Gasoline emerges as the most commonly used fuel type, while midgrade and diesel are the least used. Interestingly, despite Gasoline's overall dominance, premium Gasoline ranks as the most frequently used, surpassing regular Gasoline, which comes in as the second most common fuel type.**

**Corelation HeadMap**

```python
df_temp = df.select_dtypes(include=['float64','int64'])

corr_matric = df_temp.corr()
sns.heatmap(corr_matric,annot=False,cmap='coolwarm')
plt.show()
```

The diagram suggests an inverse relationship between engine displacement, the number of cylinders, and both the fuel efficiency score and greenhouse gas score. This indicates that as the engine size and the number of cylinders increase, the vehicle tends to become less efficient in terms of fuel economy and emissions scores.

Additionally, it shows that engine displacement and the number of cylinders are positively correlated with CO2 emissions, meaning that larger engines generally produce higher levels of emissions.
